##### 1. **WSGI 是什麼？**

* WSGI（Web Server Gateway Interface，Web伺服器閘道介面）是一種針對Python語言設計的標準介面，定義了Web伺服器與Web應用程式（或框架）之間如何進行溝通。它並不是一個軟體、模組或框架，而是一個規範，讓不同的Web伺服器和應用程式可以根據這個標準互相溝通，達到解耦的效果。WSGI標準詳細規定了一個可呼叫的應用程式物件，該物件會接受兩個參數：一個是包含請求資訊的字典（environ），另一個是負責發起回應的函數（start_response）。

##### 2. **flask 是 WSGI server 嗎？**

* Flask本身是一個WSGI應用程式，也就是說，它符合WSGI規範，可以與任何支援WSGI的伺服器搭配使用。Flask內建了一個簡單的WSGI伺服器（由Werkzeug提供），這讓開發者在開發階段可以直接運行Flask應用程式，無需額外設置。然而，這個內建的WSGI伺服器主要用於開發和測試，不建議在正式環境下使用。

##### 總結來說：

* WSGI是什麼？

    WSGI是Python語言中Web伺服器與Web應用程式之間溝通的標準介面，它是一個規範，不是軟體或框架。

* Flask是WSGI server嗎？

    Flask不是一個獨立的WSGI伺服器，而是一個WSGI應用程式。Flask內建了Werkzeug這個簡單的WSGI伺服器，供開發測試用，但在正式環境建議使用Gunicorn、uWSGI等專業WSGI伺服器。

#### 什麼是postgre資料庫？

* PostgreSQL（簡稱 Postgre）是一個功能完整且開放原始碼的物件關聯式資料庫管理系統（ORDBMS），起源於美國加州大學柏克萊分校，由 Michael Stonebraker 所主導開發。它具備極高的彈性、穩定性與完整性，並廣泛支援大多數標準 SQL 語法，是目前全球最受歡迎的開源資料庫之一。

##### 主要特色

- **開源自由**

PostgreSQL 以開放原始碼許可證發行，任何人都可以免費下載、使用、修改與散布，無論是個人、商業或學術用途。

- **物件關聯式架構**

除了傳統關聯式資料庫功能外，PostgreSQL 還整合物件導向特性，支援自訂資料型別、函數、運算子、聚合函數、索引方法與程序式語言，讓使用者能根據需求擴充資料庫功能。

- **高度擴充性**

使用者可以自由新增資料型別、函數、運算子、索引方法與程序式語言，也能安裝各式外掛模組（如地理空間處理的 PostGIS、時序資料庫支援等），大幅提升應用彈性。

- **支援多平台**

PostgreSQL 可在 Windows、Linux、macOS 等主流作業系統上執行，適用於單機、雲端、資料倉儲、Web 服務等多種應用場景。

- **強大的資料完整性與事務處理**

支援 ACID（不可部分完成性、一致性、隔離性、持久性）事務，具備外來鍵、觸發器、可更新檢查表、多版本並行控制（MVCC）等進階功能，確保資料安全與一致性。

- **豐富的資料型別與索引**

支援多種原生資料型別（如布林、整數、浮點數、字串、日期時間、JSONB 等），並提供多種索引類型（如 B+ Tree、雜湊、GIN、GiST、Sp-GiST）與全文檢索功能。

- **廣泛的語言與協定支援**

支援多種程式語言（如 C、C++、Python、Java、Go、.NET、Ruby、Perl、Tcl 等）與協定（如 ODBC），開發者可用熟悉的語言進行開發。


##### 與其他資料庫的比較

| 特性 | PostgreSQL | MongoDB（NoSQL） |
| :-- | :-- | :-- |
| 資料儲存單元 | 資料列（元組） | JSON 文件 |
| 資料結構 | 預定義結構描述（資料表） | 無結構（集合） |
| 索引方式 | B+ Tree、雜湊、GIN 等 | B Tree、化合物、雜湊等 |
| 資料一致性 | 高度一致 | 最終一致（預設） |
| 擴充性 | 高（可自訂、外掛模組） | 高（橫向擴充） |
| 適用場景 | 結構化資料、事務處理 | 非結構化、大規模資料 |

##### 應用場景

PostgreSQL 適用於多種產業與應用，包括金融、製造、零售、物流、地理資訊系統（GIS）、資料倉儲、Web 服務等。其強大的資料一致性、豐富的功能與優秀的擴充性，使其成為開發者與企業的首選資料庫之一。

#### 社群與生態系

PostgreSQL 擁有活躍的全球開發社群與龐大的生態系，包括核心開發團隊與第三方貢獻者，持續推動資料庫功能創新與最佳化。使用者可獲得社群支援、豐富的文件與各式工具，解決開發與維運上的各種問題。

##### 總結

PostgreSQL 是一個功能強大、穩定可靠、高度擴充且開放原始碼的物件關聯式資料庫管理系統，適合各類結構化資料應用與進階資料管理需求。

### 5. 使用 python 連線至 PostgreSQL server

1. 於 `lesson7` 資料夾中，建立一個 `lesson7_1.ipynb` ，在 ipynb 裡面做測試。

    1. 將需要的套件名字加入到 `requirements.txt` 裡面，在**專案的根目錄**新增終端機執行以下程式碼，便會自動安裝在 ```requirements.txt``` 裡面所列的套件。

        ```
        pip install -r requirements.txt
        ```

    2. 撰寫 Python 程式碼連線 PostgreSQL
    
        ```python

        import psycopg

        conn_string = "External Database URL" # External Database URL 即 render 的 External Database URL 完整資訊
        conn = psycopg.connect(conn_string)

        print("connection established")     
        ```
   
        <br>以下為程式碼的詳細說明：<br>

        1. `import psycopg`

            - 這行是匯入 Python 的 `psycopg` 模組（psycopg3），它是 Python 用來連接並操作 PostgreSQL 資料庫的官方驅動程式。<br>
            - 透過這個模組，你可以用 Python 程式與 PostgreSQL 伺服器溝通，執行 SQL 查詢、更新資料等。

        2. `conn_string = "postgresql://..."`

            - 這行是定義一個 **連線字串（connection string）**，用來描述如何連接到 PostgreSQL 資料庫。<br>
            - 格式為：

                ```
                postgresql://<使用者名稱>:<密碼>@<主機名稱>/<資料庫名稱>
                ```

            - 這串字串告訴 `psycopg` 要用哪個帳號、密碼，連到哪台主機的哪個資料庫。

        3. `conn = psycopg.connect(conn_string)`

            - 這行使用 `psycopg` 提供的 `connect()` 函數，透過剛剛定義的連線字串，建立一個與 PostgreSQL 資料庫的連線物件。<br>
            - 連線成功後，`conn` 會是一個連線實例，後續可以用它來建立游標（cursor）執行 SQL 查詢、交易控制等。<br>
            - 若連線失敗，程式會拋出錯誤。

        4. `print("connection established")`

            - 這行在連線建立成功後，印出提示訊息「connection established」，表示程式已成功連上 PostgreSQL 資料庫。<br>
            - 方便開發者確認連線狀態。

        5. 總結

            這段程式碼的主要作用是：

            1. 匯入 PostgreSQL 的 Python 連接套件 `psycopg`。<br>
            2. 利用一組完整的連線字串資訊，建立與遠端 PostgreSQL 資料庫的連線。<br>
            3. 連線成功後，印出提示訊息，讓使用者知道連線已經建立。

        這是 Python 與 PostgreSQL 互動的第一步，後續可以用 `conn` 建立游標來執行查詢、插入、更新等資料庫操作。
